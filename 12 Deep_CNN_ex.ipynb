```
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import Flatten, Conv2D, MaxPooling2D
from keras.utils import np_utils
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import time
```

```
# 데이터 불러오기
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```
# 입력 데이터 정규화
X_train_norm = X_train.astype('float32') / 255
X_test_norm = X_test.astype('float32') / 255
# 1행 출력
print(X_train_norm[0])
```

```
# 출력 데이터의 원-핫 인코딩 [5]
y_train_class = np_utils.to_categorical(y_train)
y_test_class  = np_utils.to_categorical(y_test)
# 1~10행 출력
print(y_train_class[0:10])
```

```
# 모형 구조
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),
                 activation='relu',
                 input_shape=(28, 28, 1)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))
```

```
model.summary()
```

```
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```
# 학습 시작 시간
start_time = time.time()

# 학습
results = model.fit(X_train_norm, y_train_class,
   validation_data = (X_test_norm, y_test_class),
   epochs=100, batch_size=256)

# 학습 종료 시간
elapsed_time = time.time() - start_time
print("경과 시간:", time.strftime("%H:%M:%S", time.gmtime(elapsed_time)))
```

```
# 학습모형에 저장된 키(변수)
print(results.history.keys())
```

```
# 학습 수에 따른 loss 변화
plt.figure(figsize=(10, 5))
plt.plot(results.history['loss'])
plt.plot(results.history['val_loss'])
plt.title('loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train', 'test'], loc='upper right')
plt.show()
```

```
# 학습 수에 따른 정확도 변화
plt.figure(figsize=(10, 5))
plt.plot(results.history['accuracy'])
plt.plot(results.history['val_accuracy'])
plt.title('accuracy')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend(['train', 'test'], loc='upper right')
plt.show()
```

```
# 모형 테스트
score = model.evaluate(X_test_norm, y_test_class)
print("Test loss:", score[0])
print("Test accuracy:", score[1])
```

```
# 예측
y_pred = model.predict(X_test_norm)
print(y_pred)
```

```
# 예측 값의 라벨
y_pred_class = np.argmax(y_pred, axis=1)
print(y_pred_class)
```

```
# 교차표: 실제 값 대비 예측 값
pd.crosstab(y_test, y_pred_class)
```
