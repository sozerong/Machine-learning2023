```python
import numpy as np
# 필요한 라이브러리를 불러옵니다.

n_data_per_class = 50
n_classes = 3
# 각 클래스당 데이터 포인트 수와 클래스 수를 설정합니다.

x = []
y = []
# 데이터와 레이블을 저장할 빈 리스트를 생성합니다.

for c in range(n_classes):
    # 클래스 수만큼 반복합니다.
    x_ = 0.3 * np.random.random_sample((n_data_per_class, 2))
    # 각 클래스당 무작위 데이터 포인트를 생성합니다.
    x_ += np.random.random_sample((1, 2))
    # 데이터 포인트에 작은 변동을 추가합니다.
    x.append(x_)
    y.append(np.asarray([c] * n_data_per_class))
    # 데이터와 레이블을 리스트에 추가합니다.

x = np.vstack(x)
y = np.concatenate(y)
# 생성된 데이터와 레이블을 NumPy 배열로 변환합니다.

from bokeh.plotting import figure, show, output_notebook
from bokeh.layouts import gridplot
from bokeh.io import export_png
# Bokeh를 사용하여 데이터 시각화를 위한 필요한 모듈을 불러옵니다.

output_notebook()
# Bokeh를 Jupyter 노트북에서 사용하기 위한 설정을 합니다.

title = '{} classes {} points'.format(n_classes, n_classes * n_data_per_class)
# 그래프 제목을 생성합니다.

def draw_figure(x, y, title, show_figure=True,
    colors = 'firebrick darksalmon lightseagreen'.split()):

    p = figure(width=400, height=400, title=title)
    # Bokeh 그래프 객체를 생성합니다.

    for c in range(n_classes):
        idx = np.where(y == c)[0]
        x_ = x[idx]
        p.scatter(x_[:,0], x_[:,1], fill_color=colors[c], line_color=colors[c])
        # 클래스별로 데이터 포인트를 색상을 다르게 하여 그래프에 표시합니다.

    if show_figure:
        show(p)
        # 그래프를 보여줍니다.

    return p
# 데이터를 Bokeh 그래프로 그리는 함수를 정의합니다.

p = draw_figure(x, y, title)
# 초기 데이터를 Bokeh 그래프로 그립니다.

from sklearn.manifold import LocallyLinearEmbedding

plots = []
# 결과 그래프들을 저장할 빈 리스트를 생성합니다.

for n_neighbors in [2, 3, 5, 10]:
    z = LocallyLinearEmbedding(
        n_components=2,
        n_neighbors=n_neighbors
    ).fit_transform(x)
    # LLE를 사용하여 데이터를 2차원으로 임베딩합니다.
    title = 'LLE with n_neighbors = {}'.format(n_neighbors)
    plots.append(draw_figure(z, y, title, show_figure=False))
    # LLE 결과를 그래프로 그리고 결과 그래프 리스트에 추가합니다.

gp = gridplot([[plots[0], plots[1]], [plots[2], plots[3]])
show(gp)
# LLE 결과 그래프를 그리고 그래프를 표시합니다.
```

```python
from sklearn.manifold import Isomap
# scikit-learn에서 Isomap을 불러옵니다.

plots = []
# 결과 그래프들을 저장할 빈 리스트를 생성합니다.

for n_neighbors in [2, 3, 5, 10]:
    z = Isomap(
        n_components=2,
        n_neighbors=n_neighbors
    ).fit_transform(x)
    # Isomap을 사용하여 데이터를 2차원으로 임베딩합니다.
    title = 'ISOMAP with n_neighbors = {}'.format(n_neighbors)
    # 그래프 제목을 생성합니다.
    plots.append(draw_figure(z, y, title, show_figure=False))
    # Isomap 결과를 그래프로 그리고 결과 그래프 리스트에 추가합니다.

gp = gridplot([[plots[0], plots[1]], [plots[2], plots[3]]])
# 결과 그래프를 그리드 형태로 배열합니다.
show(gp)
# 그리드 그래프를 표시합니다.
```
