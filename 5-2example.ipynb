```python
import pandas as pd
file = '/content/drive/MyDrive/Colab Notebooks/Online Retail.xlsx'
df = pd.read_excel(file)
df
```

```python
#Apriori 알고리즘
from mlxtend.frequent_patterns import apriori, association_rules
import matplotlib.pyplot as plt
```

```python
# dataframe 으로 변환
df[df['Quantity'] <= 0]
```

```python
df = df[df.Quantity>0]
df
```

```python
#거래별, 상품명별 수량 합계
df = df.groupby(['InvoiceNo', 'Description'])['Quantity'].sum()
df
```

```python
# 데이터 프레임의 재구조화
df=df.unstack()
df = df[:2000]
```

```python
# 전달되는 x 값들을 조건에 따라 변환하는 함수 정의
def encode_units(x):
    if x >= 1:
        return 1
    else:
        return 0

# 데이터프레임(df)에 대해 encode_units 함수를 적용하여 모든 요소를 변환
df = df.applymap(encode_units)

# 변환된 데이터프레임 출력
df
```

```python
# 데이터프레임(df)에서 각 열(axis=0)의 빈도를 합산하여 'Frequency'라는 열 이름을 가진 새로운 데이터프레임 생성
freq = df.sum(axis=0).to_frame('Frequency')

# 'Frequency' 열을 기준으로 빈도를 내림차순으로 정렬한 새로운 데이터프레임 생성
freq_sort = freq.sort_values('Frequency', ascending=False)

# 빈도가 내림차순으로 정렬된 데이터프레임 출력
freq_sort
```

```python
# 'Frequency' 열의 값을 전체 빈도 수로 나눈 'Support' 열 추가
freq_sort['Support'] = freq_sort['Frequency'] / len(freq_sort)

# 'Support' 열이 추가된 데이터프레임 출력
freq_sort
```

```python
freq_sort['Support'].hist()
```

```python
freq_items = apriori(df,min_support=0.01, use_colnames=True)
freq_items
```

```python
rules = association_rules(freq_items, metric="confidence", min_threshold=0.6)
rules
```

```python
rules=rules[rules['lift']>1]
rules.sort_values(by='confidence', ascending=False)
```

```python
rules[rules.antecedents.apply(lambda x: 'HERB MARKER THYME' in x)]
```

```python
rules[rules.consequent.apply(lambda x: 'HERB MARKER THYME' in x)]
```
