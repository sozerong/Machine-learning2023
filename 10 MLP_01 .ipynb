```python
from sklearn.neural_network import MLPClassifier
import numpy as np
```

```python
# XOR 연산을 위한 데이터 작성
# 입력값
X = np.array([[0,0],[0,1],[1,0],[1,1]])
# 목표값
y = np.array([0,1,1,0])
```

```python
# 모형화
model = MLPClassifier(hidden_layer_sizes=(2),
                      activation='logistic',
                      solver = 'lbfgs',
                      max_iter=100)
```

```python
#학습
model.fit(X,y)
```

```python
#예측
print(model.predict(X))
```
