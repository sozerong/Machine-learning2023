```python
from sklearn.linear_model import Perceptron
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 입력 값
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
# 목표 값
y = np.array([0, 1, 1, 1])
```

```python
# 산포도
plt.scatter(X[:, 0],  X[:, 1], c=y)
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
#학습 모형
model = Perceptron(verbose=1)
#학습
model.fit(X,y)
```

```python
# 가중치
print(model.coef_)
# 편향
print(model.intercept_)
# 학습 수(epoch)
print(model.n_iter_)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlabel('x1')
plt.ylabel('x2')

#파라미터
w1 = model.coef_[0,0]
w2 = model.coef_[0,1]
b = model.intercept_

# x절편
x_intercept = -b/w1
# y절편
y_intercept = -b/w2
# 선형 분류자 추가
plt.plot([0, y_intercept], [x_intercept, 0], c='red')
plt.show()
```

```python
# 테스트 입력 데이터
X_test = X
# 예측
pred = model.predict(X_test)
print(pred)
```
