```python
from sklearn import svm
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 학습용 데이터(기존 갤체)
# 입력
X_train = np.array([[2,3], [1,2]])
# 라벨
y_train = np.array([1,-1])
# 테스트 데이터
# 입력
X_test = np.array([[3,3]])
```

```python
#산포도
#학습용 데이터
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)
# 테스트용 데이터
plt.scatter(X_test[:, 0], X_test[:, 1], c='red', marker='D', s=100)
plt.xlabel('x1')
plt.ylabel('x2')
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.show()
```

```python
# SVM 분류 모형화
clf = svm.SVC(kernel='linear')
# 모형 학습
clf.fit(X_train,y_train)
```

```python
# 분류별 서포트 벡터의 수
print(clf.n_support_)
```

```python
# 서포트 벡터
print(clf.support_vectors_)
```

```python
# 서포트 벡터의 색인
print(clf.support_)
```

```python
# 각 개체들의 색인별 분류
print(clf.classes_)
```

```python
# w : 가중치
print(clf.coef_)
```

```python
# b: 편향
print(clf.intercept_)
```

```python
# 테스트 데이터의 분류
y_pred = clf.predict(X_test)
print(y_pred)
```
